```python
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.applications.efficientnet import preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np, matplotlib.pyplot as plt

# 1. Load pretrained EfficientNet
model = EfficientNetB0(weights='imagenet')

# 2. Chuẩn bị ảnh
img = image.load_img("/content/drive/MyDrive/image/Valid/Dogs/31.jpg", target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

# 3. Dự đoán
preds = model.predict(x)
label = decode_predictions(preds, top=1)[0][0]

# 4. Hiển thị
plt.imshow(image.load_img("/content/drive/MyDrive/image/Valid/Dogs/31.jpg"))
plt.title(f"EfficientNetB0 predicts: {label[1]} ({label[2]*100:.2f}%)")
plt.axis("off")
plt.show()
```
