```python
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
import numpy as np, matplotlib.pyplot as plt
import tensorflow as tf

# 1. Load model pretrained
model = ResNet50(weights='imagenet')

# 2. Tải ảnh trực tiếp từ URL
url = "https://vienmoitruong5014.org.vn/wp-content/uploads/2023/03/anh-cho-con-de-thuong_022907461.jpg"
img_path = tf.keras.utils.get_file("pug1.jpg", origin=url)

# 3. Đọc và xử lý ảnh
img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)

# 4. Dự đoán
preds = model.predict(x)
label = decode_predictions(preds, top=1)[0][0]

# 5. Hiển thị
plt.imshow(image.load_img(img_path))
plt.title(f"ResNet50 predicts: {label[1]} ({label[2]*100:.2f}%)")
plt.axis("off")
plt.show()
```
